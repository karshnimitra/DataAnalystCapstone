## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
query2 = """SELECT Age, WorkWeekHrs FROM master"""

AgeWWH = pd.read_sql_query(query2,conn)
AgeWWH.plot(kind="scatter",x="Age",y="WorkWeekHrs")
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

query3 = """SELECT Age, WorkWeekHrs, CodeRevHrs FROM master"""

AgeWWH = pd.read_sql_query(query3,conn)
AgeWWH.plot(kind="scatter",x="CodeRevHrs",y="WorkWeekHrs",s="Age")
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

query4 = """SELECT DatabaseDesireNextYear,
Count(DatabaseDesireNextYear) as Count FROM DataBaseDesireNextYear Group by DatabaseDesireNextYear
ORDER BY Count(DatabaseDesireNextYear) DESC
LIMIT 5"""

piedata = pd.read_sql_query(query4,conn)

piedata
```

```python
piedata.plot(kind="pie",y="Count",labels=piedata["DatabaseDesireNextYear"],figsize=(20,10),autopct='%1.2f%%')
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

query5 ="""SELECT Age, avg(WorkWeekHrs),
avg(CodeRevHrs) FROM master WHERE Age BETWEEN 30 and 35
GROUP BY AGE   
ORDER BY AGE"""

stackdata = pd.read_sql_query(query5,conn)
stackdata
```

```python
stackdata.plot(kind="bar",stacked=True,x="Age",figsize=(20,10))
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

query6 ="""SELECT Age, avg(ConvertedComp)
FROM master WHERE Age BETWEEN 45 and 60
GROUP BY AGE   
ORDER BY AGE"""

linedata = pd.read_sql_query(query6,conn)
plt.plot(linedata["Age"],linedata["avg(ConvertedComp)"])
```
